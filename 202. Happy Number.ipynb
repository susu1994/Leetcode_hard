当出现相同数字的时候，意味着进入循环，retrun False,不存在infinity 因为9999999999999->1053，
两种检查是否进入循环的方法：
1.set 检查是否存在，不存在，存入，存在，return false
2.快慢指针 如果快的追上慢的，说明存在循环
小tips： 每一位数字不用//10这样循环，直接转成str然后遍历每一个char

```python
class Solution:
    def isHappy(self, n: int) -> bool:
        hashset = set()
        while n!=1:
            if n in hashset:
                return False
            hashset.add(n)
            n=sum([int(i)**2 for i in str(n)])
        return True
    
```

```python
test = Solution()
test.isHappy(64)
```
